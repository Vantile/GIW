Observar que la información puede contener alguna inconsistencia.Se pide realizar funciones en Python que permitan realizar las siguientes acciones:

* Usando los dos archivos indicados, se pide crear un nuevo archivo csv denominados __NoTerrazas.csv__ el cual contenga la información de los locales que no son terrazas. Es decir, contendrá la misma información que el archivo __Locales.csv__ pero sin la información de locales que son Terrazas.

* Usando el archivo __Locales.csv__ se pide crear una nuevo archivo csv denominada __Agrupadas.csv__ el cual debe contener una línea por cada barrio conteniendo el número de locales en los que el nombre del vial del edificio en que se encuentran es diferente del nombre del vial de acceso al local. Por ejemplo el local con id 10000401 se encuentra en la Plaza de Isabel II pero el acceso al mismo se realiza desde la Calle Arenal. La información en el archivo debe aparecer en el siguiente orden:

  Código Barrio, Nombre Barrio, Número de empresas cumpliendo la condición
  


* Usando el archivo __Terrazas.csv__ crea un nuevo archivo cvs denominado TopRepeticiones.cvs. Considera el número de veces que se repite el nombre de una terraza (campo rótulo). Se pide crear un archivo csv que contenga los 20 nombres que más se repiten. La información en el archivo debe aparecer en orden descendente con los siguientes campos de información:

  Nombre terraza, Número de repeticiones

Para leer los archivos usa un código como el siguiente:

```python
import csv

#Punto 1
#Una vez ejecutado el método,para comprobar que funciona, buscar en terrazas y locales el id:10001586.Y existe en ambos,
#en cambio en noTerrazas.csv no esta, por lo funciona correctamente.
def noTerrazas():
    #Abrimos el archivo locales
    csvarchivoLocales = open('Locales.csv',encoding="utf8",errors='ignore')
    entradaLocales = csv.reader(csvarchivoLocales, delimiter=";")
    
    #Abrimos el archivo Terrazas
    csvarchivoTerrazas = open('Terrazas.csv',encoding="utf8",errors='ignore')
    entradaTerrazas = csv.reader(csvarchivoTerrazas, delimiter=";")
    #recorremos el lector y guardamos el id de terrazas en una lista
    terrazas=[]
    for linea in entradaTerrazas:
        #quitamos la primera linea
        if entradaTerrazas.line_num == 1:
            continue
        idTerraza = str(linea[0])  
        terrazas.append(idTerraza)

    #abrimos el archivo CSV para escribir en NoTerrazas.csv
    archivoWrite = open("NoTerrazas.csv", "w")
    salidaEscrito = csv.writer(archivoWrite, delimiter=';', lineterminator='\n')
    
    #Recorremos el archivo Locales
    for linea in entradaLocales:
        #quitamos la primera linea
        if entradaLocales.line_num == 1:
            continue       
        idLocal = str(linea[0])
        #vemos si existe en terrazas, la lista que contiene los id de las terrazas
        #si no existe lo insertanos en el archivo de salida
        if idLocal not in terrazas:  
            salidaEscrito.writerow(linea)
        else:
            continue
    archivoWrite.close()

#llamada a la funcion
noTerrazas()
```

```python
import csv

#Punto 2
def agrupadas():
    # Apertura del archivo de lectura
    csvarchivo = open('Locales.csv',encoding="utf8",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    filas = 0
    # Lista de salida con los valores que cumplen la condicion
    listaSal = []
    # Lista de entrada con los valores del archivo
    listaEnt = list(entrada)
    # Añade a la lista de salida las filas que cumplen la condicion del ejercicio
    for row in listaEnt:
        if filas == 0: # La primera fila se ignora
            filas = 1
            continue
        if listaEnt[filas][15] != listaEnt[filas][24] or listaEnt[filas][16] != listaEnt[filas][25]:
            listaAux = []
            listaAux.append(listaEnt[filas][3])
            listaAux.append(listaEnt[filas][4])
            listaSal.append(listaAux)
        filas = filas + 1 # Para avanzar en la lista de entrada
    filas = 0
    nombres = dict() # Diccionario para los nombres de los barrios; la clave es el codigo del barrio
    salida = dict() # Diccionario para el numero de empresas que cumplen la condicion
    # Empareja nombre y codigo en un diccionario y actualiza el numero de empresas que
    # cumplen la condicion en el otro diccionario.
    for row in listaSal:
        aux = listaSal[filas][0]
        nombre = listaSal[filas][1]
        nombres[aux] = nombre
        if aux in salida:
            num = salida[aux]
            salida[aux] = num + 1
        else:
            salida[aux] = 1
        filas = filas + 1
    # Cierre del archivo de lectura
    csvarchivo.close()
    # Apertura del archivo de escritura
    archSalida = open('Agrupadas.csv',"w")
    writer = csv.writer(archSalida, delimiter=';', lineterminator='\n')
    # Lista de codigos de los barrios que cumplen la condicion
    listaKeys = nombres.keys()
    primeraFila = True
    # Escribe en el archivo Agrupadas.csv el resultado de la funcion
    for cod in listaKeys:
        if primeraFila: # En la primera fila se pone la informacion del archivo
            writer.writerow(["Codigo Barrio","Nombre Barrio","Numero empresas"])
            primeraFila = False
            continue
        listAux = []
        listAux.append(cod)
        listAux.append(nombres[cod])
        listAux.append(salida[cod])
        writer.writerow(listAux)
    archSalida.close() # Cierre del archivo de escritura

#llamada a la funcion
agrupadas()
```

```python
import csv

#Punto 3
def topRepeticiones():
    #Abrimos el archivo Terrazas
    csvarchivo = open('Terrazas.csv',encoding="utf8",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    #clave=valor de la terrza, valor= numero de veces que se repite en el archivo
    diccionario = dict()
    
    #recorremos el archivo
    for linea in entrada:
        #quitamos la primera linea
        if entrada.line_num == 1:
            continue

        rotulo = str(linea[21])
        #vamos acumulando en el diccionario o creamos un nuevo valor si este no existia antes
        if rotulo in diccionario:
            diccionario[rotulo] += 1
        else:
            diccionario[rotulo] = 1

    #abrimos el archivo CSV para escribir
    archivoWrite = open("TopRepeticiones.csv", "w")
    salidaEscrito = csv.writer(archivoWrite, delimiter=';', lineterminator='\n')

    # ordenamos la lista
    lista = list(reversed(sorted(diccionario.items(), key=lambda x:x[1])))

    primeraFila = True
    #recorremos la lista y vamos guardando los 20 más contaminantes
    for linea in lista[:20]:
        if primeraFila: # En la primera fila se pone la informacion del archivo
            salidaEscrito.writerow(["Nombre Terraza","Numero repeticiones"])
            primeraFila = False
    
        salidaEscrito.writerow(linea)

    archivoWrite.close()

#llamada a la funcion
topRepeticiones()
```

Considerar el archivo __Agenda.json__ que contiene información de actividades de carácter gratuito, para público infantil y adulto, desarrolladas por las siguientes instituciones de titularidad municipal: la red de Bibliotecas Públicas, la Biblioteca Histórica, la Biblioteca Musical Víctor Espinós, la Hemeroteca y la Imprenta. Se presentan las actividades de los próximos 60 dias. Se pide realizar funciones en Python que permitan realizar las siguientes acciones:

* El archivo contiene un campo denominado "audience" que indica el público al que va dirigido. A veces aparece y otras veces no aparece dicho campo. Se quiere procesar el archivo para obtener como resultado dos archivos de texto:el archivo __Niños.txt__ que contendrá la información de los eventos que es exclusivamente para niños y el archivo __Familias.txt__ que contendrá la información de los eventos que está dirigida a niños y familias. En los archivos debe aparecer una línea por cada evento cumpliendo las condiciones indicadas, y mostrando la información siguiente:
  
   Título del evento(campo title), inicio(campo "dtstart"), finalizacion(campo "dtend"), url(campo link)

De manera que se quiere crear una función que tomando como entrada una longitud, una latitud y un número real que representa una distancia en kilometros, devuelva como resultado los eventos que se encuentran a una distancia menor o igual a la distancia dada como parámetro. El resultado se mostrará por la pantalla mostrando una línea por cada evento. De cada evento se mostrará:

```python
import json
leer = json.loads(open('Agenda.json',encoding="utf8").read())

#Listas donde guardaremos los eventos de cada tipo
niños = []
familia = []

#Leemos todos los eventos del fichero, accediendo a cada uno de los documentos del documento "@graph"
for i in leer["@graph"]:
    if "audience" not in i: #Comprobamos que el documento tenga el campo "audience"
        continue
    else:
        if "Niños" in i["audience"] and "Familia" not in i["audience"]: #Si contiene exclusivamente la palabra "Niños" lo añadimos
            niños.append("Título del evento: " + i["title"] + ", " + "Inicio: " + i["dtstart"] + ", " + "Finalización: " + i["dtend"] + ", " + "Url: " + i["link"])

        elif "Familia" in i["audience"]: #Si contiene la palabra "Familia" (no exclusivamente) lo añadimos
            familia.append("Título del evento: " + i["title"] + ", " + "Inicio: " + i["dtstart"] + ", " + "Finalización: " + i["dtend"] + ", " + "Url: " + i["link"])
            
#Pasamos a escribir los resultados en dos nuevos ficheros
niñosSalida = "Niños.txt" #Nombre del fichero
arc_write_n = open(niñosSalida, 'w') #Abrimos el fichero en modo escritura

#Escribimos en el fichero Niños.txt los eventos correspondientes, que serán cada elemento de la lista niños
#Además los mostramos por pantalla para comprobar el resultado
print("ARCHIVO Niños.txt:")
for lineaN in niños:    
    arc_write_n.write(lineaN + '\n')
    print(lineaN)
    
arc_write_n.close() #Cerramos el fichero de escritura

#Hacemos lo mismo con el fichero Familia.txt
familiaSalida = "Familia.txt" #Nombre del fichero
arc_write_f = open(familiaSalida, 'w') #Abrimos el fichero en modo escritura

print("----------------------------------------------------------------") #Sepración entre los dos ficheros
print("ARCHIVO Familia.txt:")
for lineaF in familia:    
    arc_write_f.write(lineaF + '\n')
    print(lineaF)

arc_write_f.close() #Cerramos el fichero de escritura
```

```python
import json
import math

#Función para calcular la distancia entre dos puntos
def haversine(lat1, lon1, lat2, lon2):
    rad=math.pi/180
    dlat=lat2-lat1
    dlon=lon2-lon1
    R=6372.795477598
    a=(math.sin(rad*dlat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*dlon/2))**2
    distancia=2*R*math.asin(math.sqrt(a))
    return distancia
#Abrimo y leemos el archivo Agenda.json
leer = json.loads(open('Agenda.json',encoding="utf8").read())
#Introducimos el punto y la distania deseada
latEntrada = input("Latitud: ")
longEntrada = input("Longitud: ")
distancia = input("Distancia: ")

#Comprobamos los eventos que están a una distancia igual o menor a la indicada y los mostramos por pantalla
for i in leer["@graph"]:
    #Hacemos cast de todos los números a float para que no los interprete como un string
    if haversine(float(i["location"]["latitude"]),float(i["location"]["longitude"]),float(latEntrada),float(longEntrada)) <= float(distancia):
        print("Título del evento: " + i["title"] + ", " + "Inicio: " + i["dtstart"] + ", " + "Finalización: " + i["dtend"] + ", " + "Url: " + i["link"]) 
```
